```python
import numpy as np
import pandas as pd
data=pd.read_csv('test_energy_data.csv')
data = data.values
print(data.shape)
```

```python
print(data)
```

```python
simple1=data[:,0]
elements1,numbers1=np.unique(simple1,return_counts=True)
print(elements1,numbers1)
```

```python
import matplotlib.pyplot as plt
labels = ['Commercial', 'Industrial', 'Residential']
sizes = [26, 34, 40]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Building Types')
```

```python
totals=[]
np.set_printoptions(precision=2)
for name in elements1:
    sum=np.sum(data[data[:,0]==name,6])
    totals.append([name,sum])
print(totals)
```

```python
x=[]
y=[]
for i in totals:
    x.append(i[0])
    y.append(i[1])
plt.bar(x,y,width=0.5)
plt.title('Consumption of Buildings')
plt.xlabel('Building Type')
plt.ylabel('Consumption')
plt.show()
```
