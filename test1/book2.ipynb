```python
import matplotlib.pyplot as plt
import numpy as np
x=[12,23,34,45,56,67,78,89]
y=[10,25,30,45,50,65,70,85]
plt.figure(figsize=(8,8))
plt.plot(x,y)
#画布大小
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
testnum=np.random.random(100)
plt.hist(testnum,bins=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=[15,15,25,25,20]
plt.pie(x,labels=['1','2','3','4','5'])
plt.show()
```
